# Tensorflow 2.0 Keras + iris

- Keras : DNN model 생성을 위한 고수준 API

```python
import tensorflow as tf 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import minmax_scale # x변수 전처리
from tensorflow.keras.utils import to_categorical # y변수 전처리
from tensorflow.keras import Sequential # model 생성
from tensorflow.keras.layers import Dense # layer 생성
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import load_model # model save/load
```

```python
# 1. x,y data
iris = load_iris()

# x : 1 ~ 4
x_data = iris.data
x_data.shape # (150,4)

x_data = minmax_scale(x_data)

# y : 5
y_data = iris.target
y_data.shape # (150,)

# reshape
y_data = y_data.reshape(-1,1)
y_data.shape # (150,1)


# [수정]
y_data = to_categorical(y_data) # y변수 전처리 : one hot encoding
y_data.shape # (150,3)


x_train, x_test, y_train, y_test = train_test_split(x_data,y_data) # 75% 25%
print(x_train.shape)
print(x_test.shape)
```

```python

# 2. keras model 생성
model = Sequential() # keras model obj
model # object info




# 3. model layer
'''
.add : model에 layer추가
model.add(Dense(node수, input_shape, activation)) # hidden layer1
model.add(Dense(node수, activation)) # hidden layer2
'''
# hidden layer1 = (4,12)
model.add(Dense(12, input_shape=(4,), activation='relu'))
# hidden layer2 = (12,6)
model.add(Dense(6, activation='relu'))
# output layer = (6,3)
model.add(Dense(3, activation='softmax'))




# 4. model compile : 학습환경 설정
model.compile(optimizer='adam', # 최적화 알고리즘(Lr 생략)
              loss='categorical_crossentropy', # 손실
              metrics=['accuracy']) # 평가방법

# layer 확인 : model info check
model.summary()
```

```python

# 5. model training : train(112) vs test(38)
model.fit(x=x_train, y=y_train, # 학습용
          epochs=300,
          verbose=1,
          validation_data=(x_test,y_test)) # 평가용



# 6. model evaluation : 모델 검증
model.evaluate(x_test,y_test) # accuracy: 0.9737
```

```python

# 7. model save / load
model.save("keras_model_iris.h5")
print("file save")


new_model = load_model("keras_model_iris.h5")
new_model.summary()



# 8. model test : new dataset
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data) 
y_pred = new_model.predict(x_test) # 예측치 : 확률값
y_true = y_test # 관측치 : one hot encode

from sklearn.metrics import accuracy_score
y_pred = tf.argmax(y_pred, axis=1)
y_true = tf.argmax(y_true, axis=1)
acc = accuracy_score(y_true, y_pred)
print("acc=",acc) # acc= 0.9736842105263158
```
