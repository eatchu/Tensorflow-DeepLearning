# 문제1)
다음 a와 b변수를 대상으로 행렬 브로드캐스팅 연산을 수행하여 출력하시오.
-    조건1> a변수 : placeholder() 이용 -> 함수 인수 대체
-    조건2> b변수 : Variable()이용
-    조건3> c변수 계산식 : c = a * b 
      -         multiply()이용 
-    조건4> a,b,c변수 결과 출력    

```python
import tensorflow as tf

a = tf.constant([1,2,3])
b = tf.Variable([4,5,6])
c = tf.multiply(a,b)
print("a={}, b={}, c={}".format(a.numpy(), b.numpy(),c.numpy()))
```

# 문제2)
텐서플로우 행렬 연산 실습

```python
x = tf.Variable([[1, 2],
                 [3, 4]])
y = tf.Variable([[1, 2, 3],
                 [2, 3, 4]])
z = tf.matmul(x,y)
print(z.numpy())
```

# 문제3)
image.jpg 이미지 파일을 대상으로 파랑색 우산 부분만 slice 하시오.

```python
import matplotlib.image as mp_image
import matplotlib.pyplot as plt 

filename = "C:/IITT/6_Tensorflow/data/image.jpg"
input_image = mp_image.imread(filename)
input_image.shape # ((432, 650, 3))
```

```python
print("********** 원본 사진 **************")
plt.imshow(input_image)
plt.show()

print("*********** 수정 사진 **************")
input_slice = tf.slice(input_image, [100,30,0],[-1,530,3])
plt.imshow(input_slice)
```
