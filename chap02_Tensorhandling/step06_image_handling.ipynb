# image slice

```python

import matplotlib.image as img  # image read
import matplotlib.pyplot as plt # image show
import tensorflow as tf

filename = "C:/IITT/6_Tensorflow/data/packt.jpeg"
input_image = img.imread(filename)

print('input dim =', input_image.ndim) #dimension -> imput dim = 3
print('input shape =', input_image.shape) #shape -> input shape = (80,144,3)

# image 원본 출력 
print("\n\n **********original photo********** \n")
plt.imshow(input_image)
plt.show() 

# image slice
img_slice = tf.slice(input_image, [15,0,0],[16,-1,-1])

print(img_slice.shape) 

# image slice 출력 
print("\n\n **********new photo********** \n")
plt.imshow(img_slice)
plt.show()


```

# image transpose

```python
import matplotlib.image as img  # image read
import matplotlib.pyplot as plt # image show
import tensorflow as tf

filename = "C:/IITT/6_Tensorflow/data/packt.jpeg"
input_image = img.imread(filename)

print('input dim =', input_image.ndim) #dimension -> imput dim = 3
print('input shape =', input_image.shape) #shape -> input shape = (80,144,3)

# image 원본 출력 
print("\n\n **********original photo********** \n")
plt.imshow(input_image)
plt.show() 

# image transpose : 축 변경[0,1,2]
img_tran = tf.transpose(a=input_image, perm=[1,0,2])
print("\n\n **********new photo********** \n")
plt.imshow(img_tran)

print(img_tran.shape) # (144,80,3)
```
