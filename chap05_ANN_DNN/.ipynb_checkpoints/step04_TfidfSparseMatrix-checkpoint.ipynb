1. csv file read
2. text, target 
3. max features
4. Sparse matrix

```python

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. csv file read
spam_data = pd.read_csv('C:/IITT/6_Tensorflow/data/temp_spam_data2.csv',header=None, encoding='utf-8' )
spam_data # 5574 x 2

target = spam_data[0] # y변수
texts = spam_data[1] # x변수

# 2. texts, target -> 전처리
# 1) target -> dummy 변수 : spam = 1, ham = 0
target = [1 if t=='spam' else 0 for t in target]
target # [0, 1, 0, 1, 0]
import pandas as pd
tar = pd.Series(target)
tar.value_counts()
'''
0    4827 -> ham
1     747 -> spam
'''

# 2) text 전처리


# 3. max features
'''
사용할 x변수의 갯수 (열의 차수)
'''
tfidf = TfidfVectorizer()
vocs = tfidf.fit(texts).vocabulary_
vocs
len(vocs) # 8722
# max_features = len(vocs) 
max_features = 4000
'''
max_features = 4000 -> 8722개 단어 중 4000개 단어만 선택
sparse matrix = [5574 x 4000]
'''


# 4. Sparse matrix

# max_features 적용 예
sparse_mat = TfidfVectorizer(stop_words='english',
                             max_features=max_features).fit_transform(texts)
sparse_mat # <5574x4000 sparse matrix of type '<class 'numpy.float64'>
print(sparse_mat)

# scipy -> numpy
sparse_mat_arr = sparse_mat.toarray()
sparse_mat_arr



# 5. train/test split
from sklearn.model_selection import train_test_split

# 70% : 30%
x_train, x_test, y_train, y_test = train_test_split(sparse_mat_arr, target, test_size = 0.3)
x_train.shape # (3901,4000)
x_test.shape # (1673,4000)



# 6. numpy binary file save
import numpy as np

spam_data = (x_train, x_test, y_train, y_test)

# allow_pickle=True
np.save('C:/IITT/6_Tensorflow/data/spam_data.npy', spam_data) # spam_data.npy
print("file save")

# file load
x_train, x_test, y_train, y_test = np.load('C:/IITT/6_Tensorflow/data/spam_data.npy',allow_pickle=True)

print(x_train.shape) # (3901, 4000)
print(x_test.shape) # (1673, 4000)
```
