 - 실제 image 적용 합성곱과 폴딩 연산

```python

import tensorflow.compat.v1 as tf # ver1.x
tf.disable_v2_behavior() # ver2.0 사용안함
from matplotlib.image import imread # image read
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. image read 
img = imread("C:/IITT/6_Tensorflow/workspace/chap07_CNN/images/parrots.png")
plt.imshow(img)

img.shape # (512,768,3)
```

```python

# 2. 실수형 변환 : int -> float32
img = img.astype('float32')
img.max() # 1.0


# 3. 정규화 : 정규화 되어 있음 0 ~ 1


# input image reshape  
img.shape # (28,28)
firstImg = img.reshape(1,512,768,3) # 합성곱을 위해 reshape : (size, h, w, color)
firstImg.shape


# Filter 변수 정의 : 사용자 지정
Filter = tf.Variable(tf.random_normal([9,9,3,8])) # 난수 (row, col, color, fmap)

```

```python


# 1. Convolution layer : 특징 추출
conv2d = tf.nn.conv2d(firstImg, Filter, strides=[1,2,2,1], padding='SAME')
print(conv2d) # shape=(1, 256, 384, 8)



# 2. Pool layer : down samplint
pool = tf.nn.max_pool(conv2d, ksize=[1,7,7,1],strides=[1,4,4,1],
            padding = 'SAME')
print(pool) # shape=(1, 64, 96, 8)




with tf.Session() as sess :
    sess.run(tf.global_variables_initializer()) # filter 초기화 
    
    # 합성곱 연산 
    conv2d_img = sess.run(conv2d)    
    conv2d_img = np.swapaxes(conv2d_img, 0, 3) # 축 교환 
    print(conv2d_img.shape) # (5, 14, 14, 1)
    
    for i, img in enumerate(conv2d_img) :
        plt.subplot(1, 8, i+1) # 1행5열,1~5 
        plt.imshow(img.reshape(256, 384), cmap='gray') # 
    plt.show()
    
    # 폴링 연산 
    pool_img = sess.run(pool)
    pool_img = np.swapaxes(pool_img, 0, 3)
    
    for i, img in enumerate(pool_img) :
        plt.subplot(1,8, i+1) # 1행5열,1~5 
        plt.imshow(img.reshape(64, 96), cmap='gray') 
    plt.show()
    
```
